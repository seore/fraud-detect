```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Allow importing from src/
PROJECT_ROOT = os.path.abspath(os.path.join(".."))
if PROJECT_ROOT not in sys.path:
    sys.path.append(PROJECT_ROOT)

from src.preprocess import load_data

sns.set(style="whitegrid")
```

```python
df = load_data(os.path.join(PROJECT_ROOT, "data", "creditcard.csv"))
df.head()
```

```python
display(df.describe().T)

print("\nShape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nNull values per column:")
print(df.isna().sum())
```

```python
class_counts = df["Class"].value_counts()
class_percent = df["Class"].value_counts(normalize=True) * 100

print("Counts:\n", class_counts)
print("\nPercentages (%):\n", class_percent)

plt.figure(figsize=(4,4))
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.xticks([0, 1], ["Non-fraud (0)", "Fraud (1)"])
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.tight_layout()

out_path = os.path.join(PROJECT_ROOT, "assets", "class_distribution.png")
plt.savefig(out_path, dpi=500, bbox_inches="tight")

plt.show()
print("Saved:", out_path)
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x="Amount", hue="Class", bins=50, log_scale=(False, True), element="step", stat="density")
plt.title("Transaction Amount Distribution (log Y)")
plt.xlabel("Amount")
plt.ylabel("Density (log scale)")
plt.tight_layout()

out_path = os.path.join(PROJECT_ROOT, "assets", "amount_distribution.png")
plt.savefig(out_path, dpi=500, bbox_inches="tight")

plt.show()
print("Saved:", out_path)
```

```python
if "Time" in df.columns:
    plt.figure(figsize=(6,4))
    sns.histplot(data=df, x="Time", hue="Class", bins=50, element="step", stat="density")
    plt.title("Transaction Time Distribution by Class")
    plt.xlabel("Time (seconds from first transaction)")
    plt.ylabel("Density")
    plt.tight_layout()
    plt.show()
```

```python
# Only numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Correlation with Class
corr_with_class = df[numeric_cols].corr()["Class"].sort_values(ascending=False)
display(corr_with_class)

# Choose top 8 features most correlated with Class (excluding Class itself)
top_features = corr_with_class.drop("Class").abs().sort_values(ascending=False).head(8).index.tolist()
print("Top correlated features:", top_features)

plt.figure(figsize=(8,6))
corr_matrix = df[top_features + ["Class"]].corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation of Top Features with Class")
plt.tight_layout()

out_path = os.path.join(PROJECT_ROOT, "assets", "corr_top_features.png")
plt.savefig(out_path, dpi=150, bbox_inches="tight")

plt.show()
print("Saved:", out_path)
```

```python
# Pick two of the top features for visualization
if len(top_features) >= 2:
    f1, f2 = top_features[0], top_features[1]
else:
    f1, f2 = "V1", "V2"  # fallback for tiny/synthetic datasets

plt.figure(figsize=(6,5))
sns.scatterplot(
    data=df.sample(min(len(df), 2000), random_state=42),
    x=f1, y=f2, hue="Class", alpha=0.6
)
plt.title(f"Scatter of {f1} vs {f2} by Class")
plt.tight_layout()
plt.show()
```
